5.Draw box plot for each variable and identify the IQR and outliers

```python
boxplot = df.boxplot(column = ['fixed acidity'])
plt.show()
```

```python
from numpy import percentile
```

```python
quartiles = percentile(df["fixed acidity"], [25,50,75])
dmin,dmax = df["fixed acidity"].min(),df["fixed acidity"].max()
```

```python
print("Min :", dmin)
print("Quartile Q1 :", quartiles[0])
print("Quartile Q2:", quartiles[1])
print("Quartile Q3:", quartiles[2])
print("Max :", dmax)
```

```python
iqr = quartiles[2]- quartiles[0]
print("Interquartile range:", iqr)
```
